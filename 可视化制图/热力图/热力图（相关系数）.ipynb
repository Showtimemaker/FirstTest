```python
import pandas as pd
data1 =pd.read_excel("C:\\Users\\周恒\\Desktop\\编号1修改1.xlsx")
data2 =pd.read_excel("C:\\Users\\周恒\\Desktop\\编号2修改1.xlsx")
data1.head()
```

```python
data2.head()
```

```python
res=pd.DataFrame(data2.values-data1.values,
                  columns=["厚度","孔隙率","压缩回弹性率","过滤阻力","过滤效率","透气性","a"])
res
```

```python
data2.describe()
```

```python
res=res.drop(columns="a")
```

```python
corr = res.corr()
corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.subplots(figsize=(12, 12)) # 设置画面大小
plt.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题,负号正常显示

# annot=True，是显式热力图上的数值；vmax是显示最大值;xticklabels、yticklabels轴标签显示；square=True，将图变成一个正方形，默认是一个矩形；cmap="Blues"是一种模式，就是图颜色配置。
# mask:控制某个矩阵块是否显示出来,默认值是None,如果是布尔型的DataFrame，则将DataFrame里True的位置用白色覆盖掉
plt.title('变量相关系数 - 差值\n', fontsize=18)

sns.heatmap(corr, annot=True, vmax=1, vmin=-1, xticklabels=True, yticklabels=True, square=True, cmap="plasma_r",
linewidths=0.05, linecolor='white') # mask=t < 0.8等价于mask=(t < 0.8)
#plt.show()

fig=plt.gcf()

fig.savefig('./666.jpg')
```
